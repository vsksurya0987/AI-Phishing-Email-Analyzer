```python
# ==========================================
# Load PDF Documents
# ==========================================

import os
from langchain_community.document_loaders import PyPDFLoader

documents = []

for file in os.listdir("knowledge"):
    if file.endswith(".pdf"):
        print(f"Loading {file}...")
        loader = PyPDFLoader(os.path.join("knowledge", file))
        documents.extend(loader.load())

print("\n✅ PDF loading completed.")
print("Total PDF pages:", len(documents))
```

```python
print(type(documents))
print(len(documents))
```

```python
all_documents = documents + [txt_document]

print("Total documents:", len(all_documents))
```

```python
# ==========================================
# Split All Documents into Chunks
# ==========================================

from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=100
)

all_chunks = text_splitter.split_documents(all_documents)

print(f"✅ Total chunks created: {len(all_chunks)}")
```

```python
from langchain_community.vectorstores import FAISS

vector_db = FAISS.from_documents(
    documents=all_chunks,
    embedding=embedding_model
)

print("✅ FAISS Database Created Successfully")
```

```python
# ==========================================
# RAG Search Function
# ==========================================

def search_knowledge(query: str, k: int = 3):
    """
    Search the FAISS knowledge base.
    """
    return vector_db.similarity_search(query, k=k)

print("✅ RAG Search Function Created Successfully")
```

```python
# ==========================================
# Import Tool
# ==========================================

import re

from crewai.tools import tool

print("✅ CrewAI Tool Imported Successfully")
```

```python
# ==========================================
# URL Extraction Tool
# ==========================================

@tool("URL Extraction Tool")
def extract_urls(email_text: str) -> str:
    """
    Extract all URLs from an email.
    """

    pattern = r"https?://[^\s]+"

    urls = re.findall(pattern, email_text)

    if not urls:
        return "No URLs Found."

    return "\n".join(urls)
```

```python
sample_email = """
Dear Customer,

Click here immediately.

https://fake-bank.xyz/login

Regards
"""

print(extract_urls.run(sample_email))
```

```python
# ==========================================
# Suspicious Keywords
# ==========================================

SUSPICIOUS_KEYWORDS = [
    "urgent",
    "verify",
    "password",
    "bank",
    "account",
    "click",
    "otp",
    "payment",
    "login",
    "security",
    "blocked",
    "suspended",
    "immediately",
]
```

```python
# ==========================================
# Risk Score Tool
# ==========================================

@tool("Risk Score Tool")
def calculate_risk(email_text: str) -> str:
    """
    Calculate phishing risk score.
    """

    score = 0

    email = email_text.lower()

    for word in SUSPICIOUS_KEYWORDS:
        if word in email:
            score += 8

    score = min(score, 100)

    if score >= 70:
        level = "HIGH"
    elif score >= 40:
        level = "MEDIUM"
    else:
        level = "LOW"

    return f"""
Risk Score : {score}%

Risk Level : {level}
"""
```

```python
print(calculate_risk.run(sample_email))
```

```python
# ==========================================
# RAG Tool
# ==========================================

@tool("Knowledge Base Search")
def rag_search(question: str) -> str:
    """
    Search the phishing knowledge base.
    """

    docs = search_knowledge(question)

    context = ""

    for doc in docs:
        context += doc.page_content + "\n\n"

    return context
```

```python
print(
    rag_search.run(
        "How can I identify phishing emails?"
    )
)
```

```python
# ==========================================
# Gemini Email Analyzer Tool
# ==========================================

@tool("Gemini Email Analyzer")
def analyze_with_gemini(email_text: str) -> str:
    """
    Analyze an email using Gemini.
    """

    prompt = f"""
You are an experienced cybersecurity expert.

Analyze the following email and provide:

1. Summary
2. Phishing Indicators
3. Social Engineering Techniques
4. Risk Explanation

Email:

{email_text}
"""

    response = client.models.generate_content(
        model="gemini-2.5-flash",
        contents=prompt
    )

    return response.text
```

```python
# ==========================================
# Import CrewAI
# ==========================================

from crewai import Agent, Task, Crew, Process

print("✅ CrewAI Imported Successfully")
```

```python
# ==========================================
# Email Analyzer Agent
# ==========================================

email_analyzer = Agent(
    role="Email Security Analyst",

    goal="Analyze phishing emails and identify phishing indicators.",

    backstory="""
You are an expert cybersecurity analyst.

Your responsibility is to analyze emails,
identify phishing indicators,
social engineering attacks,
urgent language,
and suspicious requests.

Always use the Gemini Email Analyzer tool.
""",

    tools=[analyze_with_gemini],

    verbose=True,

    allow_delegation=False
)

print("✅ Email Analyzer Agent Created")
```

```python
# ==========================================
# URL Inspector Agent
# ==========================================

url_inspector = Agent(
    role="URL Inspection Specialist",

    goal="Extract all URLs and identify suspicious links.",

    backstory="""
You specialize in identifying phishing URLs,
fake banking domains,
and malicious login pages.
""",

    tools=[extract_urls],

    verbose=True,

    allow_delegation=False
)

print("✅ URL Inspector Agent Created")
```

```python
# ==========================================
# Risk Assessment Agent
# ==========================================

risk_assessor = Agent(
    role="Risk Assessment Specialist",

    goal="Calculate phishing risk score.",

    backstory="""
You evaluate phishing emails
and determine the overall phishing risk.
""",

    tools=[calculate_risk],

    verbose=True,

    allow_delegation=False
)

print("✅ Risk Assessment Agent Created")
```

```python
# ==========================================
# Recommendation Agent
# ==========================================

recommendation_agent = Agent(
    role="Cybersecurity Advisor",

    goal="Provide recommendations using the knowledge base.",

    backstory="""
You educate users about phishing attacks,
safe browsing,
and email security.

Always search the knowledge base before answering.
""",

    tools=[rag_search],

    verbose=True,

    allow_delegation=False
)

print("✅ Recommendation Agent Created")
```

```python
# ==========================================
# AI Manager (Updated)
# ==========================================

def ai_manager(email_text):

    print("📧 Running Email Analyzer...")
    analysis = analyze_with_gemini.run(email_text)

    print("🔗 Extracting URLs...")
    urls = extract_urls.run(email_text)

    print("⚠️ Calculating Risk...")
    risk = calculate_risk.run(email_text)

    print("📚 Searching Knowledge Base...")
    context = rag_search.run(email_text)

    recommendation_prompt = f"""
You are an expert Cybersecurity Analyst.

Analyze the following information and generate a professional phishing report.

==============================
EMAIL ANALYSIS
==============================
{analysis}

==============================
DETECTED URLS
==============================
{urls}

==============================
RISK SCORE
==============================
{risk}

==============================
KNOWLEDGE BASE CONTEXT
==============================
{context}

Generate the report in the following format:

📧 Email Summary

⚠️ Risk Level

🔗 Suspicious URLs

🚩 Reasons for Detection

✅ Recommendations

🛡️ Best Practices

Use clear and simple language.
"""

    # Gemini API Call (New SDK)
    response = client.models.generate_content(
        model="gemini-2.5-flash",
        contents=recommendation_prompt
    )

    return response.text


print("✅ AI Manager Ready")
```

```python
# ==========================================
# Main Analysis Function
# ==========================================

def analyze_email(email_text):

    if not email_text.strip():
        return "⚠️ Please enter an email."

    result = ai_manager(email_text)

    return result

print("✅ Analysis Function Ready")
```
